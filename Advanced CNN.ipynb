```
import tensorflow as tf
from tensorflow.keras import layers,models
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt

# Load data
(xtr,ytr),(xte,yte)=cifar10.load_data()
xtr,xte=xtr/255.0,xte/255.0
ytr,yte=to_categorical(ytr,10),to_categorical(yte,10)

# Models
lenet=lambda:models.Sequential([
 layers.Conv2D(6,5,activation='relu',input_shape=(32,32,3)),layers.AveragePooling2D(),
 layers.Conv2D(16,5,activation='relu'),layers.AveragePooling2D(),
 layers.Flatten(),layers.Dense(120,activation='relu'),
 layers.Dense(84,activation='relu'),layers.Dense(10,activation='softmax')])

alex=lambda:models.Sequential([
 layers.Conv2D(96,3,activation='relu',input_shape=(32,32,3)),layers.MaxPooling2D(),
 layers.Conv2D(256,3,activation='relu'),layers.MaxPooling2D(),
 layers.Conv2D(384,3,activation='relu'),layers.Conv2D(256,3,activation='relu'),
 layers.MaxPooling2D(),layers.Flatten(),
 layers.Dense(1024,activation='relu'),layers.Dense(10,activation='softmax')])

vgg=lambda:models.Sequential([
 layers.Conv2D(64,3,padding='same',activation='relu',input_shape=(32,32,3)),
 layers.Conv2D(64,3,padding='same',activation='relu'),layers.MaxPooling2D(),
 layers.Conv2D(128,3,padding='same',activation='relu'),
 layers.Conv2D(128,3,padding='same',activation='relu'),layers.MaxPooling2D(),
 layers.Flatten(),layers.Dense(512,activation='relu'),
 layers.Dense(10,activation='softmax')])

def resnet():
 i=layers.Input((32,32,3))
 x=layers.Conv2D(64,3,padding='same',activation='relu')(i)
 for _ in range(3):
  s=x
  x=layers.Conv2D(64,3,padding='same',activation='relu')(x)
  x=layers.Conv2D(64,3,padding='same')(x)
  x=layers.Add()([x,s]); x=layers.Activation('relu')(x)
 x=layers.GlobalAveragePooling2D()(x)
 return models.Model(i,layers.Dense(10,activation='softmax')(x))

models_dict={"LeNet":lenet(),"AlexNet":alex(),"VGG16":vgg(),"ResNet":resnet()}
acc={}

# Train & evaluate
for n,m in models_dict.items():
 m.compile('adam','categorical_crossentropy',['accuracy'])
 m.fit(xtr,ytr,epochs=5,batch_size=64,verbose=0)
 acc[n]=m.evaluate(xte,yte,verbose=0)[1]

print(acc)

# Plot
plt.bar(acc.keys(),acc.values());plt.ylim(0,1);plt.title("CNN Comparison");plt.show()
```
