```python
import numpy as np

x = np.array([1, 2])
y = 1

w = np.random.rand(2)
b = np.random.rand()
learning_rate = 0.1

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

for i in range(100):
    z = np.dot(w, x) + b
    output = sigmoid(z)
    error = output - y
    d_sigmoid = output * (1 - output)
    w = w - learning_rate * error * d_sigmoid * x
    b = b - learning_rate * error * d_sigmoid

print("Final output:", sigmoid(np.dot(w, x) + b))
```

```python
# 1(b) Basic Neural Network with Fixed Bias Value

import numpy as np

x = np.array([1, 2])
y = 1

w = np.random.rand(2)
b = 1.0
learning_rate = 0.1

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

for i in range(100):
    z = np.dot(w, x) + b
    output = sigmoid(z)
    error = output - y
    d_sigmoid = output * (1 - output)
    w = w - learning_rate * error * d_sigmoid * x

print("Final output:", sigmoid(np.dot(w, x) + b))
```

```python
# 2) Different Activation Functions

import numpy as np
import matplotlib.pyplot as plt

def threshold(n):
    return np.where(n >= 0, 1, 0)

def sigmoid(n):
    return 1 / (1 + np.exp(-n))

def relu(n):
    return np.maximum(0, n)

def tanh(n):
    return np.tanh(n)

n = np.random.uniform(-10, 10, 100)
n = np.sort(n)

plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(n, threshold(n))
plt.title("Threshold")
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(n, sigmoid(n))
plt.title("Sigmoid Function")
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(n, relu(n))
plt.title("ReLU Function")
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(n, tanh(n))
plt.title("Tanh Function")
plt.grid(True)

plt.tight_layout()
plt.show()
```
